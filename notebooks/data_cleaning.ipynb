### data cleaning

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

There are no duplicate rows in the dataset

```python
unique_values = data.apply(lambda x: x.nunique())
for column, values in unique_values.items():
    print(f"Unique values in '{column}': {values}\n")
```

```python
num_cols = [col for col in data.columns if data[col].dtype != 'O']
cat_cols = [col for col in data.columns if col not in num_cols]
print(f'Numerical columns: {num_cols}')
print(f'Categorical columns: {cat_cols}')
```

```python
data.describe()
```

- EmployeeCount and StandardHours have only one unique value, making them uninformative for analysis or modeling. These columns can likely be dropped.

- Over18: Contains only one unique value (1). This column is redundant and can be dropped.


```python
data.drop(['StandardHours', 'Over18', 'EmployeeCount','EmployeeNumber'] , axis=1 ,inplace=True)
data.head()
```

```python
data["Attrition"] = data["Attrition"].replace({"No":0,"Yes":1})
data.head()
```
